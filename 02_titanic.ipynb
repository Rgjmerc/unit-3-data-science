# Titanic Tabulation

The Titanic was a British luxury passenger liner that sank on April 14â€“15, 1912, during its maiden voyage, en route to New York City from Southampton, England, killing about 1,500 passengers and ship personnel.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_df = pd.read_csv("titanic_passengers.csv")
```

## How many passengers were in each class?

```python
pclass = titanic_df["Pclass"].value_counts().sort_index()
plt.figure(facecolor='plum')
ax = plt.gca()
ax.set_facecolor('lavenderblush')
pclass.plot.bar(rot = 360, color="palevioletred", edgecolor = 'black')
```

  ## What were the ages of all passengers


```python
age = titanic_df["Age"]
plt.figure(facecolor='c')
ax = plt.gca()
ax.set_facecolor('powderblue')
age.plot.hist(bins = 8 , color="dodgerblue", edgecolor = "black", xlabel = "Age")
```

## Fare Cost Vs. Survival Rate

```python
fare_survival = titanic_df.groupby('Fare')['Survived'].mean().reset_index()
fare_survival.plot.scatter(x='Fare', y='Survived')
```

## Did gender affect survival rate

```python
survived = titanic_df.groupby("Sex")["Survived"].sum()
died = titanic_df.groupby("Sex")["Survived"].count() - survived
gender_survival = pd.DataFrame({"Survived": survived, "Died": died})
plt.figure(facecolor='firebrick')
gender_survival.plot.bar(color=["darkgreen","crimson"] , edgecolor="black", rot=360, ylabel="Count")
ax = plt.gca()
ax.set_facecolor('mediumseagreen')
```
