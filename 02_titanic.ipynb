```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_df = pd.read_csv("titanic_passengers.csv")
```

## How many passengers were in each class?

```python
pclass = titanic_df["Pclass"].value_counts().sort_index()
plt.figure(facecolor='lightblue')
ax = plt.gca()
ax.set_facecolor('lavenderblush')
pclass.plot.bar(rot = 360, edgecolor = 'black')
```

  ## What were the ages of all passengers


```python
age = titanic_df["Age"]
plt.figure(facecolor='c')
ax = plt.gca()
ax.set_facecolor('silver')
age.plot.hist(bins = 8 , edgecolor = "black", xlabel = "Age")
```
